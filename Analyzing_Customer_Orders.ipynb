**3. Analyze customer orders**

• Use a loop to calculate the total amount each customer spends

• If the total purchase value is above Rs 6,00,000 classify the customer as a high-value buyer

• If it is between Rs 5,00,000 to Rs 3,00,000 , classify the customer as a moderate buyer

• If it is below Rs 2,00,000 classify them as a low-value buyer

```python
# Spendings of each customer
customer_spending = {}

for customer, product, price, category in Orders:
  if customer not in customer_spending:
    customer_spending[customer] = 0
  customer_spending[customer] += price

for customer, spending in customer_spending.items():
  print(f"{customer}: {spending}")
```

```python
# Analyzing customer orders

customer_classification = {}

for customer, total in customer_spending.items():
    if total > 200000:
        customer_classification[customer] = "High-value Buyer"
    elif total >= 100000:
        customer_classification[customer] = "Moderate Buyer"
    else:
        customer_classification[customer] = "Low-value Buyer"

print("Customer Classification:")
for customer, classification in customer_classification.items():
    print(f"{customer}: {classification}")
```
